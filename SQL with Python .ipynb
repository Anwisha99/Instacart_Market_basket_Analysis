```python
import pandas as pd
import sqlite3
```

```python
# Load multiple CSVs
order = pd.read_csv('/content/sample_data/SQL DATA/orders.csv')
order_product = pd.read_csv('/content/sample_data/SQL DATA/order_products__train.csv')
aisle = pd.read_csv('/content/sample_data/SQL DATA/aisles.csv')
department = pd.read_csv('/content/sample_data/SQL DATA/departments.csv')
product = pd.read_csv('/content/sample_data/SQL DATA/products.csv')

# Connect to SQLite database (in-memory database for temporary use)
conn = sqlite3.connect(':memory:')

# Store in SQLite
order.to_sql('order', conn, index=False, if_exists='replace')
order_product.to_sql('order_product', conn, index=False, if_exists='replace')
aisle.to_sql('aisle', conn, index=False, if_exists='replace')
department.to_sql('department', conn, index=False, if_exists='replace')
product.to_sql('product', conn, index=False, if_exists='replace')
```

# **Top 5 popular product, aisle and department**



```python
#get the product wise order count and replace {answer2} to get the top 5 products based on the order count

query = '''
Select
product_name,
count(order_id) as order_count
from 'order_product' as A
left join 'product' as B on A.product_id = B.product_id
group by product_name
order by order_count desc
limit 5
'''

result = pd.read_sql_query(query, conn)
result
```

```python
#get the aisle wise order count and unique product,count for top 5 aisle based on the order count

query = '''
Select
C.aisle,
count(distinct a.product_id) as ordered_product_count,
count(distinct A.order_id) as order_count
from 'order_product' as A
left join 'product' as B on A.product_id = B.product_id
left join 'aisle' as C on B.aisle_id = C.aisle_id
group by aisle
order by order_count desc
Limit 5
'''

result = pd.read_sql_query(query, conn)
result
```

```python
#get the department wise aisle count,order count and unique product count for top 5 department on order count

query = '''
Select
d.department,
count(distinct c.aisle_id) as aisle_count,
count(distinct B.product_id) as ordered_product_count,
count(distinct A.order_id) as order_count

from 'order_product' as A
left join 'product' as B on A.product_id = B.product_id
left join 'aisle' as C on B.aisle_id = C.aisle_id
left join 'department' as D on b.department_id = D.department_id
group by d.department
order by order_count desc
limit 5
'''

result = pd.read_sql_query(query, conn)
result
```

# **Customer Purchase Behaviour**

```python
#get the average number of days between orders for all customers.

query = '''
Select avg(days_since_prior_order) As avg_order_days from 'order'
'''

result = pd.read_sql_query(query, conn)
result
```

```python
#get the number of irregular customers who have not ordered for more than 15 days.

query = '''
Select count(user_id) As irregular_user_count
from 'order'
where days_since_prior_order >15
'''

result = pd.read_sql_query(query, conn)
result
```

```python
#get the number of customers who have ordered between 7 to 14 days.

query = '''
Select count(user_id) as customer_count
from 'order'
where days_since_prior_order between 7 and 14
'''

result = pd.read_sql_query(query, conn)
result
```

```python
#get the number of customers who have ordered within 7 days or took more than 15 days.

query = '''
select
count(user_id) as customer_count
from 'order'
where days_since_prior_order <=7 or days_since_prior_order >15
'''


result = pd.read_sql_query(query, conn)
result
```

```python
#get the repeat customers.

query = '''

Select  count(distinct user_id) as repeat_customer
from 'order'
where order_number >1
'''

result = pd.read_sql_query(query, conn)
result
```

```python
#get the top 10 orders where number of items added to cart is more than 100

query = '''

Select order_id,
sum(add_to_cart_order) as item_added_to_cart

from order_product
group by order_id
having sum(add_to_cart_order) >100
order by item_added_to_cart desc
limit 10

'''

result = pd.read_sql_query(query, conn)
result
```

```python
#get the order frequency wise user count

query = '''
Select
Case
when days_since_prior_order <= 7 then "0-7 days"
when days_since_prior_order <= 15 then "8-15 days"
when days_since_prior_order <= 30 then "16-30 days"

end as order_frequency,
count(distinct user_id) as user_count

from 'order'

group by order_frequency

'''

result = pd.read_sql_query(query, conn)
result
```
